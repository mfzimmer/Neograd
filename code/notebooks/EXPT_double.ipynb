```python
nrows = 1
ncols = 4
plt.figure(1, figsize=(6*ncols, 5*nrows))

size_title = 25
size_label = 20

idx = 1
# log(rho) --------------------------------
plt.subplot(nrows, ncols, idx)
plt.plot( np.log10(d_res_Neo["rho"]), label="NeogradM")
plt.plot( np.log10(d_res_Neo_dbl["rho"]), label="NeogradM_dbl")
plt.title(r'$\log_{10}\rho$', fontsize= size_title)
plt.legend(loc = 'lower left')
plt.xlabel(r'iterations', fontsize= size_label)
plt.axhline(y=np.log10(pm.g_rhomin), color='red', linewidth=0.5)
plt.axhline(y=np.log10(pm.g_rhomax), color='red', linewidth=0.5)

# log(cost) --------------------------------
idx += 1
plt.subplot(nrows, ncols, idx)
plt.plot( np.log10(d_res_Neo["cost"]), label="NeogradM")
plt.plot( np.log10(d_res_Neo_dbl["cost"]), label="NeogradM_dbl")
plt.title(r'$\log_{10}$cost', fontsize= size_title)
plt.legend(loc = 'lower left')
plt.xlabel(r'iterations', fontsize= size_label)

# log(dist) --------------------------------
idx += 1
plt.subplot(nrows, ncols, idx)
plt.plot( np.log10(d_res_Neo["dist"]), label="NeogradM")
plt.plot( np.log10(d_res_Neo_dbl["dist"]), label="NeogradM_dbl")
plt.title(r'$\log_{10}$dist', fontsize= size_title)
plt.legend(loc = 'lower left')
plt.xlabel(r'iterations', fontsize= size_label)

# log(alpha) --------------------------------
idx += 1
plt.subplot(nrows, ncols, idx)
plt.plot( np.log10(d_res_Neo["alpha"]), label="NeogradM")
plt.plot( np.log10(d_res_Neo_dbl["alpha"]), label="NeogradM_dbl")
plt.title(r'$\log_{10}\alpha$', fontsize= size_title)
plt.legend(loc = 'lower left')
plt.xlabel(r'iterations', fontsize= size_label)

#plt.savefig(PATH_FIGS + "/double-Beale-4plots.png")
plt.show()
```
